```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"D:\Downloads\archive (5)\Amazon.csv")

# Show first 5 rows
df.head()
```

```python
# Check how many missing values are present in each column
df.isnull().sum()
```

```python
# Check how many duplicate rows exist in the dataset
df.duplicated().sum()
```

```python
# Remove rows that contain missing values
df = df.dropna()
```

```python
#Remove duplicate records to keep data clean
df = df.drop_duplicates()
```

```python
# Check dataset shape after cleaning
df.shape
```

```python
# Convert OrderDate column into datetime format
df["OrderDate"] = pd.to_datetime(df["OrderDate"])
```

```python
# Create Year column from OrderDate
df["Year"] = df["OrderDate"].dt.year
```

```python
# Create Month column from OrderDate
df["Month"] = df["OrderDate"].dt.month
```

```python
df.head()
```

```python
# Calculate total revenue from all orders
total_revenue = df["TotalAmount"].sum()

print("Total Revenue:", total_revenue)
```

```python
# Group data by Month and calculate total sales for each month
monthly_sales = df.groupby("Month")["TotalAmount"].sum()

# Display monthly revenue values
monthly_sales
```

```python
import matplotlib.pyplot as plt   # Library for creating charts

# Create figure size for better visibility
plt.figure(figsize=(8,5))

# Plot monthly sales values
plt.plot(monthly_sales)

# Add labels and title
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Monthly Sales Trend")

# Display the chart
plt.show()
```

```python
# Group data by ProductName and calculate total sales for each product
top_products = (
    df.groupby("ProductName")["TotalAmount"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

# Display top 5 products
top_products
```

```python
plt.figure(figsize=(8,5))

# Create bar chart for top products
top_products.plot(kind="bar")

plt.xlabel("Product Name")
plt.ylabel("Revenue")
plt.title("Top 5 Best Selling Products")

plt.show()
```

```python
# Group data by product category and calculate total sales
category_sales = (
    df.groupby("Category")["TotalAmount"]
    .sum()
    .sort_values(ascending=False)
)

# Display category wise sales values
category_sales

```

```python
plt.figure(figsize=(7,7))

# Create pie chart to show revenue share of each category
category_sales.plot(kind="pie", autopct="%1.1f%%")

plt.title("Revenue Share by Category")

plt.show()
```

```python
# Calculate average amount spent per order
average_order_value = df["TotalAmount"].mean()

print("Average Order Value:", round(average_order_value, 2))
```

```python
# Group sales by order date
daily_sales = df.groupby("OrderDate")["TotalAmount"].sum()

# Find the day with maximum sales
best_day = daily_sales.idxmax()
best_day_amount = daily_sales.max()

print("Highest Sales Day:", best_day)
print("Sales on that day:", best_day_amount)
```
